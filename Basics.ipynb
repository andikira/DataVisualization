This repository objective is to visualize the recent grads to get insight about unemployment from various major

First we import pandas and matplotlib library and don't forget to type %matplotlib inline to show the chart

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

read csv file to recent_grads dataframe 

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
raw_data_count = recent_grads.shape[0]
```

```python
recent_grads.shape
```

We want to remove row with null value we use .dropna()

```python
recent_grads = recent_grads.dropna()
```

```python
recent_grads.shape
```

```python
recent_grads.columns
```

To build a scatter chart we can use .plot() instance with several parameter, we want to build chart that axis is sample size and y is median

```python
recent_grads.plot(x='Sample_size',y='Median',title='Sample size vs median', kind = 'scatter')
```

With the same way we can make sample size vs unemployment rate, scatter chart

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind = "scatter")
```

For get insight about data sample, we can use histogram with bins parameters as range per bar and range for the axis  

```python
recent_grads["Sample_size"].hist(bins=20, range=(0,1000))
```

We can build a scatter and histogram at same time with two or more variable using scatter_matrix 

```python
from pandas.plotting import scatter_matrix 
scatter_matrix(recent_grads[['Sample_size','Median']], figsize=(9,9))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']])
```

We can see at the top 10 major which is the high share women

```python
recent_grads[:10].plot.bar(x = 'Major', y='ShareWomen',legend = False)
```
